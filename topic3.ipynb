14. Remove Duplicates: Write a Python function remove_duplicates(lst) that removes all duplicates from the list lst.

```python
def remove_duplicates(lst,parameter):
    if parameter==1:
        sec_list = []
        for i in lst:
            if i not in sec_list:
                sec_list+=[i]
        return sec_list
    elif parameter==2:
        return list(set(lst))

def random_list(start,stop,amount):
    from random import randint
    return [randint(start,stop) for i in range(amount)]

lst1 = random_list(1,10,5)

print(remove_duplicates(lst1,1))
print(remove_duplicates(lst1,2))
```

19. List Intersection: Write a function list_intersection(lst1, lst2) that returns the intersection of two lists lst1 and lst2.

```python
list1 = random_list(1,15,10)
list2 = random_list(1,10,10)

def list_intersection(list1,list2):
    sec_list = []
    for i in list1:
        for j in list2:
            if i==j:
                sec_list += [i]
    return sec_list


print(list1,list2)
print(list_intersection(list1, list2))
```

24. Find Index: Write a function find_index(lst, x) that finds the index of the first occurrence of an element x in a list lst. The function should return -1 if x is not in lst.

```python
def find_index(lst,x):
    flag=False
    for i in range(len(lst)):
        if lst[i]==x:
            save=i
            flag=True
            break
    if flag:
        return save
    else:
        return -1


lst1 = random_list(1,7,30)
print(lst1)
print(find_index(lst1,5))
print(find_index(lst1,8))
```

27. Split List: Write a function split_list(lst, n) that splits the list lst into n sublists of nearly equal size.

```python
def split_list(lst, n):
  size = (len(lst) // n)+1
  return list(map(lambda x: lst[x * size:x * size + size], list(range(n))))

print(split_list([1, 2, 3, 4, 5, 6, 7], 4))
```

33. Group Anagrams: Write a function group_anagrams(words) that groups anagrams together from the list of strings words.

```python
def group_anagrams(str1,str2):
    str1 = str1.lower()
    str2 = str2.lower()
    if(len(str1) == len(str2)):
        if(sorted(str1) == sorted(str2)):
            print(str1 + " and " + str2 + " are anagram.")
        else:
            print(str1 + " and " + str2 + " are not anagram.")
    else:
        print(str1 + " and " + str2 + " are not anagram.")

a, b = input('input: ').split() # input like silent listen

group_anagrams(a,b)
```
