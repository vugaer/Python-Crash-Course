```python
item_prices = {'apple': 1.00, 'banana': 0.50}
cem=0
new_values = []

for values in item_prices.values():
    cem+=values
    new_values += [values*0.9]

def replace_values(dict1, lst):
    keys = list(dict1.keys())

    for i in range(len(lst)):
        dict1[keys[i]]=lst[i]
    return dict1

print('Total price: %.2f\nUpdate dict: %s'%(cem, replace_values(item_prices, new_values)))
```
