**First Question**

````python
your_list = "A Z E R B A I J A N".split() # i am vetenperver btw

for i in your_list:
    print(i, end='')
````

**Second Question**

````python
from math import pi

r = float(input('Radius: '))

print('Area of circle ~%.2f'%(pi*(r**2)))
````

**Third Question**

````python
my_repo = []

for i in ['first', 'second', 'third']:
    my_repo += [int(input('Enter the %s number: '%(i)))]

for i in sorted(my_repo, reverse=True):
    print(i, end='')
````

**Fourth Question**

````python
"""please insert numbers like: 1 2 3 4 5 to get 1 4 9 16 25"""

print([i**2 for i in map(int, input().split())])
````

**Fifth Question**

````python
my_list = [i for i in map(int, input().split())]
print([sum(my_list[:i+1]) for i in range(len(my_list))])
````
